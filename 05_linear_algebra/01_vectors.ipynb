# Vectors in Python

```python
# import the liberaries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Let's create a vector
vector = np.array([5, 8])
vector
```

```python
# Let's plot this vector
plt.figure(figsize = (10, 6))
plt.quiver(0 , 0, vector[0], vector[1], angles = 'xy', scale_units = 'xy', scale = 1, color = 'blue')
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.grid()
plt.title('2D Vector')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```python
# Two vectors
v1 = np.array([1, 6])
v2 = np.array([4, 5])
```

```python
# Let's plot this vectors
plt.figure(figsize = (10, 6))
plt.quiver(0 , 0, v1[0], v1[1], angles = 'xy', scale_units = 'xy', scale = 1, color = 'blue')
plt.quiver(0 , 0, v2[0], v2[1], angles = 'xy', scale_units = 'xy', scale = 1, color = 'r')
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.grid()
plt.title('2D Vector')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
