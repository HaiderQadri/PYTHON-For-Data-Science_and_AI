## Interqurtile Range (IQR = Q3 - Q1)
    Robust for dealing outliers

```python
# generating a datadet
np.random.seed(0)
data = np.random.normal(100, 20, 200)
data
```

```python
data = np.append(data, [500, 10])
data
```

```python
# calculate IQR
Q3 = np.percentile(data, 75)
Q1 = np.percentile(data, 25)
IQR = Q3 - Q1
IQR
```

```python
# Defining crietria for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
lower_bound, upper_bound
```

```python
# checking outliers
outliers = (data < lower_bound) | (data > upper_bound)
```

```python
# removing outliers
data_with_no_outliers = data[~outliers]
data_with_no_outliers
```

```python
# ploting both the data
plt.figure(figsize = (12, 5))
plt.subplot(1, 2, 1)
sns.boxplot(data = data)
plt.title("data with outliers")

plt.subplot(1, 2, 2)
sns.boxplot(data = data_with_no_outliers)
plt.title('data without outliers')

plt.show()
```

## standard error (standard error = standard deviation / squareroot of length of the data )

```python
 # generate the dataset
data = np.array([2,3,4,5,6])
data
```

```python
std = np.std(data, ddof = 1)
std
```

```python
se = std / np.sqrt(len(data))
se
```
