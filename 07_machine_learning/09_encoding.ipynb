## Feature encoding (Using Scikit-learn liberary)

```python
# import liberaries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the DataFrame
df = sns.load_dataset('tips')
df
```

```python
df['time'].value_counts()
```

## LabelEncoder

```python
# importing liberary for encoding
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder, OneHotEncoder
```

```python
# encoding time column in df using LabelEncoder
# call LabelEncoder
le = LabelEncoder()
```

```python
# fitting tha data in to encoder
df['encoded_time'] = le.fit_transform(df['time'])
df.head(90)
```

```python
df['encoded_time'].value_counts()
```

## OrdinalEncoder

```python
df.head()
```

```python
df['day'].value_counts()
```

```python
# encoding day column using OrdinalEncoder
# calling encoder
oe = OrdinalEncoder(categories = [['Thur', 'Fri', 'Sat', 'Sun']])
```

```python
# fitting the data
df['encoded_day'] = oe.fit_transform(df[['day']])
df.head()
```

```python
df['encoded_day'].value_counts()
```

## OneHotEncoder

```python
df.head()
```

```python
df['sex'].value_counts()
```

```python
# calling encoder
ohe = OneHotEncoder()
```

```python
# fitting data 
ohe.fit_transform(df[['sex']]).toarray()
```

```python
# converting into pandas DataFrame
encoded_sex = pd.DataFrame(ohe.fit_transform(df[['sex']]).toarray(), columns=['male', 'female'])
encoded_sex
```

```python
# merging both the data
df = pd.concat([df.reset_index(drop = True), encoded_sex.reset_index(drop = True)], axis = 1)
df
```

```python
# trying another example of OneHotEncoder
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic['embarked'].value_counts()
```

```python
# calling OneHotEncoder
ohe = OneHotEncoder()
```

```python
# fitting the data into encoder
encoded_embarked = ohe.fit_transform(titanic[['embarked']]).toarray()
encoded_embarked
```

```python
# converting into pandas DataFrame
encoded_embarked_df = pd.DataFrame(encoded_embarked, columns = ohe.get_feature_names_out(['embarked']))
encoded_embarked_df
```

```python
# merging both the data frames
titanic = pd.concat([titanic.reset_index(drop = True), encoded_embarked_df.reset_index(drop = True)], axis=1)
titanic.head()
```
