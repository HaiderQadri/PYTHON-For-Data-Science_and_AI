## Visualization using seaborn liberary

```python
# import the liberaries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the dataset
df = sns.load_dataset('tips')
```

```python
df.info()
```

```python
df.head()
```

## Scatter Plot

```python
plt.figure(figsize = (12, 6))
sns.scatterplot(data = df, x = 'total_bill', y = 'tip', hue = 'sex', legend = True) # Usually size of A4 paper (width = 8, height = 11) and presentation (16, 9)
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Scatter plot of Total Bill vs Tip based on Gender')
plt.legend(loc = 'upper left', title = 'SEX')

# to save as file
# plt.savefig('./path/folder/file_name.png', dpi = 300) # Most of the time dpi = 300 
plt.show()
```

- We can set the manual colors of our palette by using dictionary 
- We can change colors
- Move the legend 
- Modify the title of legend
- Nodify values of our axis 
- Make a title of figure

```python
plt.figure(figsize = (12, 6))
palette = {'Male': 'black', 'Female': '#700f7d'}
sns.scatterplot(data = df, x = 'total_bill', y = 'tip', hue = 'sex', legend = True, palette = palette) 
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Scatter plot of Total Bill vs Tip based on Gender')
plt.legend(loc = 'upper left', title = 'SEX')

plt.show()
```

- Changing theme style 
- Set the font size manually
- Change the font style

```python
plt.figure(figsize = (12, 6))
sns.set_theme(style = 'darkgrid', font_scale = 0.8, font = 'Times New Roman')
palette = {'Male': 'black', 'Female': 'green'}
sns.scatterplot(data = df, x = 'total_bill', y = 'tip', hue = 'sex', legend = True, palette = palette)
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Scatter plot of Total Bill vs Tip based on Gender')
plt.legend(loc = 'upper left', title = 'SEX')

plt.show()
```

- Here, we changed the size of points on the basis of 'size of people'

```python
plt.figure(figsize = (12, 6))
sns.set_theme(style = 'darkgrid', font_scale = 0.8, font = 'Times New Roman')
palette = {'Male': 'black', 'Female': 'green'}
sns.scatterplot(data = df, x = 'total_bill', y = 'tip', hue = 'sex', legend = True, palette = palette, size = 'size')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Scatter plot of Total Bill vs Tip based on Gender')
plt.legend(loc = 'upper left', title = 'SEX')

plt.show()
```

## Box Plot

```python
plt.figure(figsize = (12, 6))
sns.set_theme(style = 'darkgrid', font_scale = 0.8, font = 'Times New Roman')
palette = {'Dinner': 'black', 'Lunch': 'yellow'}
sns.boxplot(data = df, x = 'sex', y = 'tip', hue = 'time', legend = True, palette = palette)
plt.xlabel('Sex')
plt.ylabel('Tip')
plt.title('Box plot of Sex vs Tip based on Time')
plt.legend(loc = 'upper left', title = 'Time')

plt.show()
```

```python
plt.figure(figsize = (12, 6))
sns.set_theme(style = 'darkgrid', font_scale = 0.8, font = 'Times New Roman')
palette = {'Dinner': 'black', 'Lunch': 'yellow'}
sns.boxenplot(data = df, x = 'sex', y = 'tip', hue = 'time', legend = True, palette = palette)
plt.xlabel('Sex')
plt.ylabel('Tip')
plt.title('Boxen plot of Sex vs Tip based on Time')
plt.legend(loc = 'upper left', title = 'Time')

plt.show()
```

## Bar plot

```python
plt.figure(figsize = (12, 6))
sns.set_theme(style = 'darkgrid', font_scale = 0.8, font = 'Times New Roman')
palette = {'Dinner': 'black', 'Lunch': 'yellow'}
sns.barplot(data = df, x = 'sex', y = 'tip', hue = 'time', legend = True, palette = palette)
plt.xlabel('Sex')
plt.ylabel('Tip')
plt.title('Bar plot of Sex vs Tip based on Time')
plt.legend(loc = 'upper left', title = 'Time')

plt.show()
```

- Draw capsize 
- Change capsize 
- Change edge color

```python
plt.figure(figsize = (12, 6))
sns.set_theme(style = 'darkgrid', font_scale = 0.8, font = 'Times New Roman')
palette = {'Dinner': 'black', 'Lunch': 'yellow'}
sns.barplot(data = df, x = 'sex', y = 'tip', hue = 'time', legend = True, palette = palette, capsize = 0.1, edgecolor = 'green')
plt.xlabel('Sex')
plt.ylabel('Tip')
plt.title('Bar plot of Sex vs Tip based on Time')
plt.legend(loc = 'upper left', title = 'Time')

plt.show()
```

## Violin Plot

```python
plt.figure(figsize = (12, 6))
sns.set_theme(style = 'darkgrid', font_scale = 0.8, font = 'Times New Roman')
palette = {'Dinner': 'black', 'Lunch': 'yellow'}
sns.violinplot(data = df, x = 'sex', y = 'tip', hue = 'time', legend = True, palette = palette)
plt.xlabel('Sex')
plt.ylabel('Tip')
plt.title('Violin plot of Sex vs Tip based on Time')
plt.legend(loc = 'upper left', title = 'Time')

plt.show()
```
