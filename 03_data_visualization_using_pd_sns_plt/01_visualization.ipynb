```python
# import liberaries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the data using seaborn liberary
df = sns.load_dataset('titanic')
df.head()
```

```python
df.head()
```

```python
# let's draw scatter plot (which is drawan between two numerical variables)
df.plot(kind = 'scatter', x = 'age', y = 'fare')
plt.title('Age vs Fare')
plt.show()
```

```python
# let's change the color of scatter points
df.plot(kind = 'scatter', x = 'age', y = 'fare', color = 'yellow') 
plt.title('Age vs Fare')
plt.show()
```

```python
# we can change colors using hex codes
df.plot(kind = 'scatter', x = 'age', y = 'fare', color = '#0c822d')
plt.title('Age vs Fare')
plt.show()
```

```python
df.head()
```

```python
# Now, let's draw bar charts (which includes one numerical and one categorical variable)
df.groupby('sex')['fare'].mean().plot(kind = 'bar', color = 'blue')
plt.show()
```

```python
df.head()
```

```python
sns.barplot(data = df, x = 'sex', y = 'fare')
plt.show()
```

```python
# we can simply flip the picture by just chaniging the axis 
sns.barplot(data = df, y = 'sex', x = 'fare')
plt.show()
```

```python
sns.scatterplot(data = df, x = 'age', y = 'fare', hue = 'class')
plt.show()
```

```python
# In the above plot we can see that two passengers are travelling in the first class who paid heighest fare.
# Now, we want to know who were those (i.e male, female etc)
sns.scatterplot(data = df, x = 'age', y = 'fare', hue = 'who')
plt.show()
```
